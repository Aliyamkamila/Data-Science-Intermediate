### Diabetes Daata set
1. age
1. sex
1.  body mass index (bmi)
1. average blood pressuare

```python
from sklearn.datasets import load_diabetes

import numpy as np
import pandas as pd
```

Target : Progression (numeric assessment of the progression of the disease in a year)

```python
data = load_diabetes()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['progreesion']=data.target
df.head()
```

```python
import matplotlib.pyplot as plt

plt.plot(df['bmi'], df['bp'], 'b.')
```

polynomial exercise 1
x : bmi
y : blood presuare

```python
x = df[['bmi']]
y = df['bp']
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_fetures = PolynomialFeatures(degree=10, include_bias=False)
x_poly = poly_fetures.fit_transform(x)
```

```python
x_poly
```

```python
from sklearn.linear_model import LinearRegression

regression = LinearRegression()
regression.fit(x_poly, y)
```

```python
from sklearn.linear_model import Ridge

regression = Ridge()
regression.fit(x_poly, y)
```

```python
from sklearn.linear_model import ElasticNet

regression = ElasticNet(alpha=2.0, l1_ratio=0.5)
regression.fit(x_poly, y)
```

```python
x_line_fitted = np.linspace(x.min(), x.max(), num=100)
y_line_fitted = regression.intercept_

for i in range(len(poly_fetures.powers_)):
  exponent = poly_fetures.powers_[i][0]
  y_line_fitted = y_line_fitted + regression.coef_[i] * x_line_fitted ** exponent
```

# Elastic Net

```python
plt.plot(x_line_fitted, y_line_fitted, 'r-')
plt.plot(x, y, 'b.')
plt.show()
```

# Ridge Regssion

```python
plt.plot(x_line_fitted, y_line_fitted, 'r-')
plt.plot(x, y, 'b.')
plt.show()
```
