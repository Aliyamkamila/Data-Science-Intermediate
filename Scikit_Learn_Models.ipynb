# Models


### Estimstor : ML Models in scikit learn
* ML Models is scikit-learn
* Fuction: Fit and Predict

```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
```

```python
import matplotlib.pyplot as plt

x, y = make_regression(n_samples=20, n_features=1, noise=3.0, random_state=42)

plt.plot(x, y, 'o')
plt.show()
```

```python
regression =  LinearRegression()
regression.fit(x, y)
```

```python
prediction = regression.predict(x)
```

```python
plt.plot(x, y, 'o')
plt.plot(x, prediction, 'r-')
plt.show()
```

MinMaxScaler: scale features between zero and one. min value will be 0 and max value will be one
