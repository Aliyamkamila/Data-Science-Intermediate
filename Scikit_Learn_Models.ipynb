# Models


### Estimstor : ML Models in scikit learn
* ML Models is scikit-learn
* Fuction: Fit and Predict

```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
```

```python
import matplotlib.pyplot as plt

x, y = make_regression(n_samples=20, n_features=1, noise=3.0, random_state=42)

plt.plot(x, y, 'o')
plt.show()
```

```python
regression =  LinearRegression()
regression.fit(x, y)
```

```python
prediction = regression.predict(x)
```

```python
regression.score(x, y)
```

```python
plt.plot(x, y, 'o')
plt.plot(x, prediction, 'r-')
plt.show()
```

# Pipelines

series of transformers, often with an estimator at the end

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline

# Membuat data regresi
x, y = make_regression(n_samples=20, n_features=1, noise=6.0, random_state=10)

# Membuat pipeline dengan Pipeline() dan menamai langkah-langkah
pipeline = Pipeline([
    ('Scale', MinMaxScaler()),
    ('Regression', LinearRegression())
])

# Melatih model
pipeline.fit(x, y)

# Prediksi
prediction = pipeline.predict(x)

# Visualisasi hasil regresi
plt.scatter(x, y, color='blue', label="Data Asli")
plt.plot(x, prediction, color='red', label="Regresi Linear")
plt.legend()
plt.show()
```

# Matrics

MSE(mean squared error): find mean squared error between target values and predicted values

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y, prediction)
```

```python
pipeline.score(x, y) #r2 score, scores clores to 1.0 are better
```
