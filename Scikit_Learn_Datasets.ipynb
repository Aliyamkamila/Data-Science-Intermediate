# Generating

```python
from sklearn.datasets import make_regression

features, targets = make_regression(n_samples=10, n_features=1, random_state=10)

features, targets
```

```python
import matplotlib.pyplot as plt

plt.plot(features, targets, 'o')
plt.show()
```
